```python
import pandas as pd
import numpy as np
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.tsa.api as smt
import matplotlib.dates as mdates
from matplotlib import cm
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
raw_data = pd.read_csv('./raw_data.csv')

data = raw_data.copy()
var_list = data.columns[2:]
for var in var_list:
    data.loc[:, var] = data.loc[:, var].str.strip("%").astype(float)

data.loc[:, 'observation_date'] = pd.to_datetime(data.loc[:, 'observation_date'], format="%Y-%m-%d")
```

```python
data.to_csv('./processed_data.csv', index=False)
```

```python
data = data.set_index('observation_date')
```

```python
print('   Total - %d Rows' % (len(data)))
for i, var in enumerate(var_list):
    num_nan = len(data[data[var].isnull()])
    print('%8s - %3d missing' % (var, num_nan))
```

```python
cut_off_year =  486
print('   Total - %d Rows' % (len(data[cut_off_year:])))
for i, var in enumerate(var_list):
    num_nan = len(data[cut_off_year:][data[cut_off_year:][var].isnull()])
    print('%8s - %3d missing' % (var, num_nan))
```

```python
fig = plt.figure(figsize=(10, 6))
ax1 = fig.add_subplot(1, 1, 1)
ax2 = ax1.twinx()

data.iloc[:, 0].plot(ax=ax1, alpha=0.6)
data.loc[:, 'GS10'].plot(ax=ax2, alpha=0.5)
ax1.legend()
ax2.legend(bbox_to_anchor=(1.08, 1), loc='upper left', ncol=1)
for tick in ax1.get_xticklabels():
    tick.set_fontsize(13)
for tick in ax1.get_yticklabels():
    tick.set_fontsize(13)
for tick in ax2.get_yticklabels():
    tick.set_fontsize(13) 
```

# Y-delta (division) vs x

```python
y_delta = (data.loc[:, 'CPIAUCNS']/data.loc[:, 'CPIAUCNS'].shift(1))-1
x = data.loc[:, 'GS10']
```

```python
fig = plt.figure(figsize=(10, 6))
ax1 = fig.add_subplot(1, 1, 1)
ax2 = ax1.twinx()

y_delta.plot(ax=ax1, alpha=0.6, color='r')
x.plot(ax=ax2, alpha=0.2)
ax1.legend()
ax2.legend(bbox_to_anchor=(1.08, 1), loc='upper left', ncol=1)
for tick in ax1.get_xticklabels():
    tick.set_fontsize(13)
for tick in ax1.get_yticklabels():
    tick.set_fontsize(13)
for tick in ax2.get_yticklabels():
    tick.set_fontsize(13) 
```

# Y-delta (division) vs x

```python
y_delta = (data.loc[:, 'CPIAUCNS']/data.loc[:, 'CPIAUCNS'].shift(12))-1
x = data.loc[:, 'GS10'].shift(-12)
```

```python
fig = plt.figure(figsize=(10, 6))
ax1 = fig.add_subplot(1, 1, 1)
ax2 = ax1.twinx()

y_delta.plot(ax=ax1, alpha=0.6, color='r')
x.plot(ax=ax2, alpha=0.2)
ax1.legend()
ax2.legend(bbox_to_anchor=(1.08, 1), loc='upper left', ncol=1)
for tick in ax1.get_xticklabels():
    tick.set_fontsize(13)
for tick in ax1.get_yticklabels():
    tick.set_fontsize(13)
for tick in ax2.get_yticklabels():
    tick.set_fontsize(13) 
```

```python
corr = y_delta.corr(x)
plt.figure(figsize=(6,6))
plt.scatter(y_delta, x, color='yellowgreen', alpha=0.3, label=('Pearson=%.3f' % corr))
plt.xlabel('CPI annual % change')
plt.ylabel('Interest Rate (GS10)')
plt.legend(fontsize=14)
```

```python
fig = plt.figure(figsize=(10, 6))
ax1 = fig.add_subplot(1, 1, 1)
ax2 = ax1.twinx()

y_delta[-240:].plot(ax=ax1, alpha=0.6, color='r')
x[-240:].plot(ax=ax2, alpha=0.2)
ax1.legend()
ax2.legend(bbox_to_anchor=(1.08, 1), loc='upper left', ncol=1)
for tick in ax1.get_xticklabels():
    tick.set_fontsize(13)
for tick in ax1.get_yticklabels():
    tick.set_fontsize(13)
for tick in ax2.get_yticklabels():
    tick.set_fontsize(13) 
```

```python
corr = y_delta[-240:].corr(x[-240:])
plt.figure(figsize=(6,6))
plt.scatter(y_delta[-240:], x[-240:], color='yellowgreen', alpha=0.3, label=('Pearson=%.3f' % corr))
plt.xlabel('CPI annual % change')
plt.ylabel('Interest Rate (GS10)')
plt.legend(fontsize=14)
```
